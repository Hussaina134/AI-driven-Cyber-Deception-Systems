```python
# week4_data_prep.ipynb
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load your Week 3 features file
df = pd.read_csv("features.csv")
print("Initial shape:", df.shape)

# Preview data
df.head()
```

```python
# Check missing values
print("Missing values per column:")
print(df.isnull().sum().sort_values(ascending=False).head(10))

# Replace NaN with 0 or column mean depending on type
for col in df.columns:
    if df[col].dtype in [np.float64, np.int64]:
        df[col] = df[col].fillna(df[col].mean())
    else:
        df[col] = df[col].fillna("unknown")

print("\n✅ Missing values handled.")
```

```python
# Drop duplicate rows if any
df = df.drop_duplicates()

# Remove obvious non-feature columns (like timestamps or IDs)
to_drop = [c for c in df.columns if "timestamp" in c.lower() or "session_id" in c.lower()]
df = df.drop(columns=to_drop, errors="ignore")

print("After cleanup:", df.shape)
```
