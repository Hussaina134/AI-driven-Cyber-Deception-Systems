```python
import pandas as pd
df = pd.read_csv("features.csv")
df.head()
```

```python
import matplotlib.pyplot as plt
plt.hist(df['duration'].dropna(), bins=40)
plt.title("Session duration distribution")
plt.xlabel("seconds")
plt.show()
```

```python
plt.hist(df['cmd_count'].fillna(0), bins=30)
plt.title("Commands per session")
plt.show()
```
